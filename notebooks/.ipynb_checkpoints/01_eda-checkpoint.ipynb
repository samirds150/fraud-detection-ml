```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../data/raw/enhanced_fraud_detection_dataset.csv')

# Overview
print("Shape of the dataset:", df.shape)
print("\nColumn Data Types:")
print(df.dtypes)

# Display top rows
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Class distribution
fraud_counts = df['fraud_flag'].value_counts()
fraud_percent = df['fraud_flag'].value_counts(normalize=True) * 100

print("Fraud Counts:\n", fraud_counts)
print("\nFraud Percentages:\n", fraud_percent)

# Plot
sns.countplot(data=df, x='fraud_flag', palette='Set2')
plt.title('Fraud vs Non-Fraud Distribution')
plt.xlabel('Fraud Flag (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
# Total and percentage of missing values
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing Percentage (%)': missing_percent
}).sort_values(by='Missing Percentage (%)', ascending=False)

print("Columns with missing values:")
missing_df[missing_df['Missing Count'] > 0]
```

```python
# Descriptive statistics for numerical columns
df.describe().T
```

```python
num_cols = ['amount', 'ip_risk_score', 'mouse_movement_score', 'session_duration_sec']

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='fraud_flag', y=col, data=df, palette='Set3')
    plt.title(f'{col} vs Fraud')
    plt.tight_layout()

plt.show()
```

```python
cat_features = ['payment_method', 'device_type', 'geo_location', 'region', 'country',
                'device_os', 'browser', 'day_of_week', 'is_vpn', 'is_international', 
                'suspicious_activity_flag', 'is_blacklisted_user']

plt.figure(figsize=(18, 25))
for i, col in enumerate(cat_features, 1):
    plt.subplot(6, 2, i)
    sns.countplot(data=df, x=col, hue='fraud_flag', palette='Set2')
    plt.title(f'{col} vs Fraud')
    plt.xticks(rotation=45)
    plt.tight_layout()
plt.show()
```

```python
num_features = ['amount', 'ip_risk_score', 'mouse_movement_score', 'account_age_days',
                'time_of_day', 'num_prev_transactions', 'avg_transaction_amount',
                'session_duration_sec', 'login_attempts_last_hour']

plt.figure(figsize=(18, 20))
for i, col in enumerate(num_features, 1):
    plt.subplot(5, 2, i)
    sns.kdeplot(data=df, x=col, hue='fraud_flag', common_norm=False)
    plt.title(f'Distribution of {col} by Fraud Flag')
    plt.tight_layout()
plt.show()
```

```python
# Select numerical features only
num_df = df.select_dtypes(include=['int64', 'float64'])

# Exclude ID columns & target
num_df = num_df.drop(columns=['user_id', 'fraud_flag'])

# Correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(num_df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

# Make a copy and encode categorical columns
df_enc = df.copy()
cat_cols = df_enc.select_dtypes(include=['object']).columns

le = LabelEncoder()
for col in cat_cols:
    df_enc[col] = le.fit_transform(df_enc[col].astype(str))

# Drop ID columns and NA values
X = df_enc.drop(columns=['transaction_id', 'user_id', 'fraud_flag']).dropna()
y = df_enc.loc[X.index, 'fraud_flag']

# Calculate mutual information
mi_scores = mutual_info_classif(X, y, discrete_features='auto')
mi_df = pd.DataFrame({'Feature': X.columns, 'MI Score': mi_scores}).sort_values(by='MI Score', ascending=False)

# Display
plt.figure(figsize=(10, 6))
sns.barplot(x='MI Score', y='Feature', data=mi_df, palette='viridis')
plt.title('Feature Importance via Mutual Information')
plt.tight_layout()
plt.show()
```
